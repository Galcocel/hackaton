```python


from transformers import pipeline

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

text = "pollo arroz ensalada verdura proteinas"
candidate_labels = ["fit", "ecologico", "dulce", "salado", "vegetariano"]

result = classifier(text, candidate_labels)
print(result["labels"])
print(result["scores"])
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.linear_model import LogisticRegression
import numpy as np

# Embeddings
model = SentenceTransformer("all-MiniLM-L6-v2")

clientes = [
    ["pollo", "arroz", "verduras"],  # fit
    ["tofu", "kale", "quinoa"],      # vegetarian
    ["galletas", "chocolate"]        # sweet
]
labels = ["fit", "vegetarian", "sweet"]

X = np.array([np.mean(model.encode(p), axis=0) for p in clientes])
y = labels

clf = LogisticRegression().fit(X, y)

# Predecir un nuevo cliente
nuevo_cliente = ["pollo", "arroz", "carne"]
X_new = np.mean(model.encode(nuevo_cliente), axis=0).reshape(1, -1)
print(clf.predict(X_new))
print(clf.predict_proba(X_new))

```
