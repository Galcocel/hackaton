```python


from transformers import pipeline

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

text = "pollo arroz ensalada verdura proteinas"
candidate_labels = ["fit", "ecologico", "dulce", "salado", "vegetariano"]

result = classifier(text, candidate_labels)
print(result["labels"])
print(result["scores"])
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.linear_model import LogisticRegression
import numpy as np

# Embeddings
model = SentenceTransformer("all-MiniLM-L6-v2")

# Ejemplo de nuevos clientes
clientes_test = [
    ["pollo", "arroz integral", "brocoli"],
    ["tofu", "kale", "quinoa"],
    ["galletas", "chocolate", "helado"]
]

# Convertir lista de alimentos en frases y generar embeddings
X_test = np.array([model.encode(" ".join(cliente)) for cliente in clientes_test])

# Predicción
preds = clf.predict(X_test)
probas = clf.predict_proba(X_test)

# Mostrar resultados con tabla
classes = clf.classes_
print("{:<10} {:>10} {:>10} {:>10} {:>10} {:>10} {:>10} {:>10}".format("Cliente", *classes))
for i, prob in enumerate(probas, start=1):
    print("{:<10} {}".format(f"Cliente {i}", " ".join([f"{p:.3f}" for p in prob])))
```
