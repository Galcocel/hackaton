```python


from transformers import pipeline

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

text = ""
candidate_labels = ["Gym Enthusiast",
    "Vegano",
    "Eco-Friendly",
    "Celíaco/Intolerancias",
    "Gourmet/Premium",
    "Conveniente/Express",
    "Familia Numerosa"]

result = classifier(text, candidate_labels)
print(result["labels"])
print(result["scores"])
```
